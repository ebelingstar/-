```python
#Два графика
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("/content/books.csv")
df23=df[df["Год"].isin([2023])]
df24=df[df["Год"].isin([2024])]

x1=df23["Месяц"]
y1=df23["Книги"]

x2=df24["Месяц"]
y2=df24["Страницы"]

def line_chart(x1, y1):
  plt.plot(x1, y1, marker="*", mfc="#4682B4", color="#FF1493")
  plt.grid(True)
  plt.xlabel("Месяцы")
  plt.ylabel("Количество книг")
  plt.title("Прочитанные книги за год")
  plt.legend("Прочитано книг за месяц")
  return plt.show()

def bar_chart(x2, y2):
  plt.bar(x2, y2, color="#98FB98")
  plt.grid(True)
  plt.xlabel("Месяцы")
  plt.ylabel("Количество страниц")
  plt.title("Прочитанные страницы за год")
  plt.legend("Прочитано страниц за месяц")
  return plt.show()

plt.subplot(2, 1, 1)
line_chart(x1, y1)
plt.subplot(2, 1, 2)
bar_chart(x2, y2)
```

```python
#Сравниение двух датасетов
import matplotlib.pyplot as plt

def two_plots(y1, y2):
  fig, ax = plt.subplots()
  ax.set_title("Сравниение двух датасетов")
  ax.plot(y1, label='Dataset 1')
  ax.plot(y2, label='Dataset 2')

  ax.set_xlabel('Индекс')
  ax.set_ylabel('Значение')
  ax.set_title('Сравнение двух датасетов')
  ax.legend()

  plt.show()
  return ax
two_plots([1, 2, 3, 4, 5], [8, 7, 6, 2, 1])
```

```python
#Круговая диаграмма
import matplotlib.pyplot as plt
import numpy as np

# Пример данных
y = np.array([35, 25, 25, 15])
fruit_labels = ["Apples", "Bananas", "Cherries", "Oranges"]

new_colors = ["#77C654", "#F9E099", "#E2445E", "#FF8C00"]

plt.pie(y, labels = fruit_labels, colors = new_colors)

plt.legend(title = "Four Fruits:")

plt.show()
```

```python
#Построение функции синуса
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots()

ax.plot(x, y, ':g')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title(' График функции синуса')

plt.show()
```
